```python
from sklearn.metrics import roc_curve
from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import PrecisionRecallDisplay
from sklearn.metrics import average_precision_score, roc_auc_score
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
%matplotlib notebook
```

```python
guideseq_df = pd.read_csv("./crisprNet_CNNstd_guideSeq_results.csv")
```

```python
sns.set_style('darkgrid')
fig, (ax, ax1) = plt.subplots(1, 2, figsize=(12,6))
fpr, tpr, _ = roc_curve(guideseq_df['label'], guideseq_df['CNN_Transformer_SMOTE'], pos_label=1)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, estimator_name='Convolutional self-attention network (AdamW)',roc_auc=roc_auc_score(guideseq_df['label'], guideseq_df['CNN_Transformer_SMOTE'])).plot(ax=ax)
fpr, tpr, _ = roc_curve(guideseq_df['label'], guideseq_df['CNN_Transformer_SMOTE_adam'], pos_label=1)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, estimator_name='Convolutional self-attention network (Adam)',roc_auc=roc_auc_score(guideseq_df['label'], guideseq_df['CNN_Transformer_SMOTE_adam'])).plot(ax=ax)
fpr, tpr, _ = roc_curve(guideseq_df['label'], guideseq_df['CRISPR_Net_Pred'], pos_label=1)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, estimator_name='CRISPR-Net',roc_auc=roc_auc_score(guideseq_df['label'], guideseq_df['CRISPR_Net_Pred'])).plot(ax=ax)
prec, recall, _ = precision_recall_curve(guideseq_df['label'], guideseq_df['CNN_Transformer_SMOTE'], pos_label=1)
fpr, tpr, _ = roc_curve(guideseq_df['label'], guideseq_df['CNN_std_Pred'], pos_label=1)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, estimator_name='CNN_std',roc_auc=roc_auc_score(guideseq_df['label'], guideseq_df['CNN_std_Pred'])).plot(ax=ax)

prec, recall, _ = precision_recall_curve(guideseq_df['label'], guideseq_df['CNN_Transformer_SMOTE'], pos_label=1)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall, estimator_name='Convolutional self-attention network (AadmW)', average_precision=average_precision_score(guideseq_df['label'], guideseq_df['CNN_Transformer_SMOTE'], pos_label=1)).plot(ax=ax1)
prec, recall, _ = precision_recall_curve(guideseq_df['label'], guideseq_df['CNN_Transformer_SMOTE_adam'], pos_label=1)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall, estimator_name='Convolutional self-attention network (Aadm)', average_precision=average_precision_score(guideseq_df['label'], guideseq_df['CNN_Transformer_SMOTE_adam'], pos_label=1)).plot(ax=ax1)

prec, recall, _ = precision_recall_curve(guideseq_df['label'], guideseq_df['CRISPR_Net_Pred'], pos_label=1)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall, estimator_name='CRISPR-Net', average_precision=average_precision_score(guideseq_df['label'], guideseq_df['CRISPR_Net_Pred'], pos_label=1)).plot(ax=ax1)
prec, recall, _ = precision_recall_curve(guideseq_df['label'], guideseq_df['CNN_std_Pred'], pos_label=1)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall, estimator_name='CNN_std', average_precision=average_precision_score(guideseq_df['label'], guideseq_df['CNN_std_Pred'], pos_label=1)).plot(ax=ax1)
ax1.legend(loc='upper right', facecolor='white')
ax.legend(loc='lower right', facecolor='white')
ax1.set_title("Precision-recall curve")
ax.set_title("Receiver operating characteristic (ROC) curve")
plt.tight_layout()
plt.savefig("./guideseq_testing.pdf", dpi=500)
```

```python
sns.set_style('darkgrid')
fig, (ax, ax1) = plt.subplots(1, 2, figsize=(12,6))
fpr, tpr, _ = roc_curve(crispor_df['transformer_alone_actuals'], crispor_df['transformer_alone_preds'], pos_label=1)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, estimator_name='CNN with self-attention', roc_auc=roc_auc_score(crispor_df['transformer_alone_actuals'], crispor_df['transformer_alone_preds'])).plot(ax=ax)
fpr, tpr, _ = roc_curve(crispor_df['transformer_alone_actuals'], crispor_df['cnn_transformer_preds'], pos_label=1)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, estimator_name='deeper CNN with self-attention', roc_auc=roc_auc_score(crispor_df['transformer_alone_actuals'], crispor_df['cnn_alone_preds'])).plot(ax=ax)
fpr, tpr, _ = roc_curve(crispor_df['transformer_alone_actuals'], crispor_df['cnn_alone_preds'], pos_label=1)
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, estimator_name='CNN with self-attention (no SMOTE)', roc_auc=roc_auc_score(crispor_df['transformer_alone_actuals'], crispor_df['cnn_transformer_preds'])).plot(ax=ax)
fpr, tpr, _ = roc_curve(crispor_df['transformer_alone_actuals'], crispor_df['transformer_noSMOTE_preds'], pos_label=1)

roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, estimator_name='CNN', roc_auc=roc_auc_score(crispor_df['transformer_alone_actuals'], crispor_df['cnn_alone_preds'])).plot(ax=ax)
prec, recall, _ = precision_recall_curve(crispor_df['transformer_alone_actuals'], crispor_df['transformer_alone_preds'], pos_label=1)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall, estimator_name='CNN with self-attention',average_precision=average_precision_score(crispor_df['transformer_alone_actuals'], crispor_df['transformer_alone_preds'], pos_label=1)).plot(ax=ax1)
prec, recall, _ = precision_recall_curve(crispor_df['transformer_alone_actuals'], crispor_df['cnn_transformer_preds'], pos_label=1)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall, estimator_name='deeper CNN with self-attention',average_precision=average_precision_score(crispor_df['transformer_alone_actuals'], crispor_df['cnn_transformer_preds'], pos_label=1)).plot(ax=ax1)
prec, recall, _ = precision_recall_curve(crispor_df['transformer_alone_actuals'], crispor_df['cnn_alone_preds'], pos_label=1)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall, estimator_name='CNN', average_precision=average_precision_score(crispor_df['transformer_alone_actuals'], crispor_df['cnn_alone_preds'], pos_label=1)).plot(ax=ax1)
prec, recall, _ = precision_recall_curve(crispor_df['transformer_alone_actuals'], crispor_df['transformer_noSMOTE_preds'], pos_label=1)
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall, estimator_name='CNN with self-attention (no SMOTE)', average_precision=average_precision_score(crispor_df['transformer_alone_actuals'], crispor_df['transformer_noSMOTE_preds'], pos_label=1)).plot(ax=ax1)
ax1.legend(loc='upper right', facecolor='white')
ax1.set_title("Precision-recall curve")
ax.legend(loc='lower right', facecolor='white')
ax.set_title("Receiver operating characteristic (ROC) curve")
plt.tight_layout()
plt.savefig("./crispor_ds_5fold_crxvalidation.pdf", dpi=500)
```
